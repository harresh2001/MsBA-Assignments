```python
import pandas as pd
import numpy as np
```

```python
# Load datasets
df1 = pd.read_csv('C:/Users/yvonn/OneDrive/Desktop/SUNWAY/202409-S4/Business Intelligence/Assignment/SocialInclusiondata.csv')
df2 = pd.read_csv('C:/Users/yvonn/OneDrive/Desktop/SUNWAY/202409-S4/Business Intelligence/Assignment/Happinessdata.csv')
```

```python
# Drop the row
df1 = df1.iloc[1:].reset_index(drop=True)
df1.head()
```

```python
# Rename the 'Country.x' column to 'Country name'
df1 = df1.rename(columns={'Country.x': 'Country name'})
df1.head()
```

```python
# Create a dictionary to map different country name variations to a standard format
country_name_corrections = {
    'Gambia, The': 'Gambia',
    "CÃ´te d'Ivoire": 'Ivory Coast',
    'Türkiye': 'Turkey',
    'Laos': 'Lao PDR',
    'Kyrgyzstan': 'Kyrgyz Republic',
    'Viet Nam': 'Vietnam',
    'Syrian Arab Republic': 'Syria',
    'Slovak Republic': 'Slovakia',
    'Bahamas, The': 'Bahamas',
    'Congo, Dem. Rep.': 'Congo (Kinshasa)',
    'Congo, Democratic Rep.': 'Congo (Kinshasa)',
    'Congo, Rep.': 'Congo (Brazzaville)',
    'Czech Republic': 'Czechia',
    'Egypt, Arab Rep.': 'Egypt',
    'Hong Kong SAR': 'Hong Kong S.A.R. of China',
    'Hong Kong SAR, China': 'Hong Kong S.A.R. of China',
    'Iran, Islamic Rep.': 'Iran',
    'Korea, Dem. People\'s Rep.': 'North Korea',
    'Korea, Rep.': 'South Korea',
    'Macao SAR, China': 'Macau',
    'Macedonia, FYR': 'North Macedonia',
    'Russia': 'Russian Federation',
    'Taiwan, China': 'Taiwan Province of China',
    'United States': 'United States of America',
    'Venezuela, RB': 'Venezuela',
    'Yemen, Rep.': 'Yemen'
}

# Replace country names in both datasets based on the mapping
df1['Country name'] = df1['Country name'].replace(country_name_corrections)
df2['Country name'] = df2['Country name'].replace(country_name_corrections)

# Replace the crossid based on corrected name
df1['crossid'] = df1['Country name'] + '_' + df1['year'].astype(str)
df2['crossid'] = df2['Country name'] + '_' + df2['year'].astype(str)
```

```python
# Merge the datasets on 'crossid' to include 'Life Ladder' from df2 in df1
merged_df = pd.merge(df1, df2[['crossid', 'Life Ladder']], 
                     on=['crossid'], how='left')
```

```python
# Convert the 'year' column to integer (if it contains non-numeric values, can use errors='coerce' to handle them)
merged_df['year'] = pd.to_numeric(merged_df['year'], errors='coerce')
```

```python
# Filter the merged dataset to include only rows where 'year' is between 2011 and 2022
filtered_df = merged_df[(merged_df['year'] >= 2011) & (merged_df['year'] <= 2022)]
```

```python
# Check for missing values in 'Nutrition & Basic Medical Care' (or other relevant columns)
filtered_df['has_value'] = filtered_df[['Nutrition & Basic Medical Care']].notna().any(axis=1)

# Sort DataFrame to ensure rows with actual values (non-NaN) are prioritized
filtered_df.sort_values(by=['crossid', 'has_value'], ascending=[True, False], inplace=True)

# Drop duplicates, keeping the first row that has non-NaN values (based on 'has_value')
cleaned_final_df = filtered_df.drop_duplicates(subset=['crossid'], keep='first')

# Drop the helper column ('has_value') that was used for sorting
cleaned_final_df.drop(columns=['has_value'], inplace=True)

# Display the cleaned DataFrame without duplicates
print(cleaned_final_df)
```

```python
# Check again duplicates
duplicates = cleaned_final_df[cleaned_final_df.duplicated(subset=['crossid'], keep=False)]
print(duplicates)
```

```python
# Extracting country name and year from 'crossid'
cleaned_final_df['Country'] = cleaned_final_df['crossid'].apply(lambda x: x.split('_')[0])
cleaned_final_df['Year'] = cleaned_final_df['crossid'].apply(lambda x: x.split('_')[1])

# Grouping by 'Country' and counting the number of rows for each
country_year_count = cleaned_final_df.groupby('Country').size()

# Filtering countries that do not have 12 rows
missing_years_countries = country_year_count[country_year_count != 12]

# Display the results
if missing_years_countries.empty:
    print("All countries have 12 rows from 2011 to 2022.")
else:
    print("The following countries do not have 12 rows from 2011 to 2022:")
    print(missing_years_countries)
```

```python
# Fix the range 2011-2022
countries_with_missing_rows = {
    'East Asia and Pacific': 6,
    'Eurasia': 6,
    'Europe and North America': 6,
    'Ivory Coast': 11,
    'Latin America and the Caribbean': 6,
    'Middle East and North Africa': 6,
    'South Asia': 6,
    'Sub-Saharan Africa': 6,
    'Swaziland': 11
}

# Define the full range of years
full_years = list(range(2011, 2023))

# Loop through each country with missing rows
for country, current_row_count in countries_with_missing_rows.items():
    # Filter the dataset for the current country
    country_data = cleaned_final_df[cleaned_final_df['Country name'] == country]
    
    # Find which years are missing
    missing_years = [year for year in full_years if year not in country_data['year'].values]
    
    # Create a dataframe for the missing rows
    missing_data = pd.DataFrame({
        'Country name': [country] * len(missing_years),
        'year': missing_years,
        'crossid': [f"{country}_{year}" for year in missing_years]  # Generate 'crossid'
    })
    
    # Add missing rows to the original dataframe
    cleaned_final_df = pd.concat([cleaned_final_df, missing_data], ignore_index=True)

# ensure 'crossid' is filled correctly for all rows with NaN
cleaned_final_df['crossid'] = cleaned_final_df.apply(
    lambda row: f"{row['Country name']}_{int(row['year'])}" if pd.isna(row['crossid']) else row['crossid'], axis=1
)

# Sort the dataframe by country name and year to maintain order
cleaned_final_df = cleaned_final_df.sort_values(by=['Country name', 'year']).reset_index(drop=True)

# Verify the result
print(cleaned_final_df[cleaned_final_df['Country name'].isin(countries_with_missing_rows.keys())])
```

```python
# Drop any additional or unnecessary columns
columns_to_drop = ['Country', 'Year']  
cleaned_final_df = cleaned_final_df.drop(columns=columns_to_drop)

# Sort the dataframe by country name and year to maintain order
cleaned_final_df = cleaned_final_df.sort_values(by=['Country name', 'year']).reset_index(drop=True)

# Verify the result
print(cleaned_final_df[cleaned_final_df['Country name'].isin(countries_with_missing_rows.keys())])
```

```python
# Check for missing values in the merged DataFrame
missing_values = cleaned_final_df.isnull().sum()

# Display the count of missing values
print(missing_values)
```
